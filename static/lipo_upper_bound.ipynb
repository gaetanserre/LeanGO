```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib
matplotlib.rcParams['text.usetex'] = True
```

```python
plt.rcParams["hatch.linewidth"] = 4
```

```python
rect_color = "#32a861"
```

```python
def gen_rects(x, x_i, up):
  rects = []
  max_x_i = np.max(np.sin(x_i))
  xs = []
  for x_ in x:
    if max_x_i <= up(x_i, [x_]):
      xs.append(x_)
    else:
      if xs == [] :
        continue
      rects.append(patches.Rectangle((xs[0], -10), xs[-1] - xs[0], 20, facecolor="white", edgecolor=rect_color, hatch=r"//", alpha=0.5))
      xs = []
  if xs != []:
    rects.append(patches.Rectangle((xs[0], -10), xs[-1] - xs[0], 20, facecolor="white", edgecolor=rect_color, hatch=r"//", alpha=0.5))
  return rects
```

```python
x = np.linspace(-np.pi-0.5, np.pi+0.5, 10_000)
y = np.sin(x)

x_i = np.random.uniform(-np.pi-0.5, np.pi+0.5, 4)

up = lambda x_i, x : [np.min(np.sin(x_i) + np.abs(x_i - x_)) for x_ in x]

rects = gen_rects(x, x_i, up)

fig, ax = plt.subplots()

for rect in rects:
  ax.add_patch(rect)
ax.add_patch(patches.Rectangle((0, 0), 0, 0, facecolor="white", edgecolor=rect_color, hatch=r"//", label=r"Potential maximizers area"))

ax.plot(x, y, linewidth=2, color="black", label="$f$")
ax.scatter(x_i, np.sin(x_i), color="navy", label="Samples")
ax.plot(x, up(x_i, x), color="#7b9ef8", label="LIPO's upper bound")
ax.set_ylim(np.min(np.sin(x))-0.2, np.max(up(x_i, x)))


#ax.grid(True, linestyle="--")
ax.set_xticklabels([])
ax.set_yticklabels([])
ax.set_xticks([])
ax.set_yticks([])

plt.legend()
plt.savefig("lipo_upper_bound.svg", bbox_inches="tight")
```
